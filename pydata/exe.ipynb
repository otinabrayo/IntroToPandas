```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
```

```python
df.head()
```

```python
import numpy as np

# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(20)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print (missing_data[column].value_counts())
    print("")    
```

```python
avr_norm_loss = df["normalized-losses"].astype(float).mean(axis=0)
df["normalized-losses"] = df["normalized-losses"].fillna(avr_norm_loss)
```

```python
avr_str = df["stroke"].astype(float).mean(axis=0)
# df["stroke"].replace(np.nan, avr_str, inplace=True);
df["stroke"] = df["stroke"].fillna(avr_str)
```

```python
avr_bor = df["bore"].astype(float).mean(axis=0)
df["bore"] = df["bore"].fillna(avr_bor)
```

```python
avr_hrs = df["horsepower"].astype(float).mean(axis=0)
df["horsepower"] = df["horsepower"].fillna(avr_hrs)
```

```python
avr_pk_rm = df["peak-rpm"].astype(float).mean(axis=0)
df["peak-rpm"] = df["peak-rpm"].fillna(avr_pk_rm)
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df["num-of-doors"] = df["num-of-doors"].fillna("four")
```

```python
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df[["bore", "stroke", "price", "peak-rpm"]] = df[["bore", "stroke", "price", "peak-rpm"]].astype("float")
df[["normalized-losses", "horsepower"]] = df[["normalized-losses", "horsepower"]].astype("int")
```

```python
df.dtypes
```

```python
df["city-mpg"] = 235/df["city-mpg"]
```

```python
df.rename(columns={"city-mpg": "city-L/100km"}, inplace=True)
```

```python
df.head(20)
```

```python
df.rename(columns={"highway-mpg": "highway-L/100km"}, inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df["horsepower"])

# set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
```

```python
group_names = ["Low", "Medium", "High"]
```

```python
df["horsepower-binned"] = pd.cut(df["horsepower"], bins, labels=group_names, include_lowest=True)
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```

```python
plt.bar(group_names, df["horsepower-binned"].value_counts())
plt.xlabel("Horsepower")
plt.ylabel("Count")
plt.title("Horsepower bins")
```
